```python
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar',figsize = (12,6), title = 'Purchase Distribution by Age')
```

```python
for i in df['Age'].unique():
    print(i,':',len(df[df['Age'] == i]))
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i,df[df['Age'] == i]['Product_ID'].nunique()])
data  = pd.DataFrame(lst,columns = ['Age','Products'])
```

```python
data
```

```python
data.plot.bar(x = 'Age', figsize = (8,6))
```

```python
df.groupby('Age').sum(numeric_only = True)['Purchase'].plot(kind = 'bar',figsize = (12,6), title = 'Amount Spend by Age')
```

```python
df.groupby('Age').mean(numeric_only = True)['Purchase'].plot(kind = 'bar',figsize = (12,6), title = 'Amount Spend by Age')
```

```python
df.groupby('Age').mean(numeric_only = True)['Purchase'].plot(kind = 'pie',autopct = "%.1f")
```
